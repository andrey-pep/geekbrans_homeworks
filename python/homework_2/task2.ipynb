```python
import pandas as pd
authors = pd.DataFrame({'author_id': [1,2,3], 'author_name': ['Тургенев', 'Чехов', 'Островский']}, columns=['author_id', 'author_name'])
authors
```

```python
books = pd.DataFrame({
    'author_id' : [1, 1, 1, 2, 2, 3, 3],
    'book_title' : ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    'price' : [450, 300, 350, 500, 450, 370, 290]
}, columns=['author_id', 'book_title', 'price'])
books
```

```python
authors_price = pd.merge(authors, books, on='author_id', how='inner')
authors_price
```

```python
ap_copy = authors_price.copy()
top5 = ap_copy.sort_values(by='price', ascending=0).head(5)
top5
```

```python
authors_max_price = authors_price.groupby('author_name').agg({'price': 'max'})
authors_max_price = authors_max_price.rename(columns={'price' : 'max_price'})
authors_max_price
authors_min_price = authors_price.groupby('author_name').agg({'price': 'min'})
authors_min_price = authors_min_price.rename(columns={'price' : 'min_price'})
authors_mean_price = authors_price.groupby('author_name').agg({'price': 'mean'})
authors_mean_price = authors_mean_price.rename(columns={'price' : 'mean_price'})

authors_stat = pd.merge(authors_max_price, authors_min_price, on='author_name', how='inner')
authors_stat = pd.merge(authors_stat, authors_mean_price, on='author_name', how='inner')
authors_stat
```
